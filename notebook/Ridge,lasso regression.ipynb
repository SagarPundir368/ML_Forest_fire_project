# Algerian Forest Fires Dataset
Data Set Information:
The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from june 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute (class) The 244 instances have been classified into fire (138 classes) and not fire (106 classes). 

## Dataset information:-
1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations 
2. Temp : temperature noon (temperature max)  in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90 
4. Ws :Wind speed in km/h: 6 to 29 
5. Rain: total day in mm: 0 to 16.8
FWI Components  
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5 
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9 
8. Drought Code (DC) index from the FWI system:  7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5 
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

### Changing columns to float data type with a more generic way

```python
object_columns = [feature for feature in df.columns if df[feature].dtypes == 'O']
object_columns
```

```python
# CONVERT ALL THE COLUMNS IN TO FLOAT EXCEPT THE CLASSES COLUMN AS IT IS A CATEGORICAL FEATURE SO WE HAVE TO DO ENCODING FOR THIS.
for i in object_columns:
    if i != 'Classes':
        df[i] = df[i].astype(float)
        
```

## **EXPLORATORY DATA ANALYSIS**

```python
# See as day,month and year is not that too important so we can drop it
df_copy = df.drop(['day','month','year'], axis=1)
df_copy.head()
```

```python
# Check the Values and their count of the Classes feature
df_copy['Classes'].value_counts()
```

### So we can see that we have same category but count many time so bascially it is due to because may be some having space after string "fire" or "fire " and "not fire" or "not fire ".
### So what we gonna try we match the string not fire or fire 

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

## MONTHLY FIRE ANALYSIS:- FOR SIDI-BEL REGION
NOW USE THE COMPLETE DATASET

```python
df.head()
```

```python
# So as we have only two category in Classes feature but some have sapca at the end or at the beginning so write
# this line so that all that thing remove and we finallly have have actually two output cateory
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),'not fire', 'fire')
```

```python
df_sidi_bel = df.loc[df['Region']==1]
df_sidi_bel.head()
```

```python
df_sidi_bel['Classes'].value_counts()
```

```python
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes', data=df_sidi_bel)
plt.ylabel("No. of fires", weight='bold')
plt.xlabel("Months",weight='bold')
plt.title("FIRES IN SIDI-BEL REGION BY MONTH")
plt.show()
```
